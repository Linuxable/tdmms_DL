## **Image Splitting comparison**

```python
from typing import Dict
import pandas as pd
import os
import matplotlib.pyplot as plt

BEP_DIR = os.path.abspath("../../")
FIGS_DIR = os.path.join(BEP_DIR, 'paper_images')
CURVES_DIR = os.path.join('bep', 'training_curves_csv')

plt.rcParams['font.family'] = 'Times New Roman'

def plot_loss_from_host(
    data: Dict[str, pd.DataFrame],
    data2: Dict[str, pd.DataFrame],
    ylim: float = None
) -> None:
    """
    Plots training curves from the 'data' dictionary created by scraping the local tensorboard host
    for data.

    Args:
        - data (dict): dictionary where the keys are the metrics and values are the corresponding dataframes.
        - ylim (float): gives the plots a y-limit.
    """

    for metric in data:
        if 'val' in metric:
            continue

        layers, loss_type = [i for i in metric.split('_') if i not in ['epoch', 'loss']]

        metric_val = metric.replace('epoch', 'epoch_val')

        layers = layers.upper()
        if loss_type != 'bbox':
            loss_type = loss_type.title()
        else:
            loss_type = 'BBOX'

        print(f'{layers} {loss_type} loss')

        plt.figure(figsize=(7, 5))
        # plt.plot(data[metric]['Step'], data[metric]['Value'], label='Train', c='#769cae')
        # plt.plot(data[metric_val]['Step'], data[metric_val]['Value'], label='Val', c='#120c0d')
        # plt.plot(data2[metric]['Step'], data2[metric]['Value'], label='Train 2', c='#769cae')
        # plt.plot(data2[metric_val]['Step'], data2[metric_val]['Value'], label='Val 2', c='#120c0d')

        plt.plot(data[metric]['Step'], data[metric]['Value'], label='Without - Train', c='red')
        plt.plot(data[metric_val]['Step'], data[metric_val]['Value'], linestyle = ':', label='Without - Val', c='red')
        plt.plot(data2[metric]['Step'], data2[metric]['Value'], label='With - Train', c='blue')
        plt.plot(data2[metric_val]['Step'], data2[metric_val]['Value'],  linestyle = ':', label='With - Val', c='blue')


        axis_label_fontsize = 15
        tick_label_fontsize = 14
        legend_fontsize = 12

        if ylim:
            plt.ylim(ylim[0], ylim[1])

        if loss_type == 'Class':
            plt.ylim(0, 0.11)


        plt.xlabel('Epoch', fontsize=axis_label_fontsize)
        plt.ylabel(f'{loss_type} Loss', fontsize=axis_label_fontsize)
        plt.xticks(fontsize=tick_label_fontsize)
        plt.yticks(fontsize=tick_label_fontsize)
        # plt.axvline(30, c='red')
        # plt.axvline(60, c='red')
        # plt.axvline(90, c='red')
        plt.legend(fontsize=legend_fontsize)

        plt.savefig(os.path.join(FIGS_DIR, f'image_splitting_comp_80_training_curve_{layers}_{loss_type}.eps'), bbox_inches='tight')
        plt.show()
    
    return None

def create_data_dict(run: str) -> Dict[str, pd.DataFrame]:
    files = [i for i in os.listdir(CURVES_DIR) if run in i]

    data = {}

    for f in files:
        data[f.replace(run, '').replace('.pkl', '')] = pd.read_pickle(os.path.join(CURVES_DIR, f))

    return data
```

```python
data_80 = create_data_dict('mos2_80_')
data_249 = create_data_dict('mos2_249_')
```

```python
plot_loss_from_host(data_80, data_249)
```
